```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
import urllib.request, urllib.error, urllib.parse
import json
import time
import random
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configure Edge WebDriver options
edge_options = Options()
edge_options.use_chromium = True  # Use Chromium-based Edge

# Set up Edge WebDriver service
edge_service = Service(EdgeChromiumDriverManager().install())

# Create an Edge WebDriver instance with the specified options
driver = webdriver.Edge(service=edge_service, options=edge_options)



brand_names=[]
product_names= []
product_ratings = []
review_count = []
product_price= []
url_list = ['https://www.amazon.com/s?k=gap&crid=2HGTPE0NNQ3OO&qid=1696181198&sprefix=gap%2Caps%2C110&ref=sr_pg_1',
            'https://www.amazon.com/s?k=gap&page=2&crid=2HGTPE0NNQ3OO&qid=1696181036&sprefix=gap%2Caps%2C110&ref=sr_pg_2',           
            'https://www.amazon.com/s?k=gap&page=3&crid=2HGTPE0NNQ3OO&qid=1696181153&sprefix=gap%2Caps%2C110&ref=sr_pg_3',
            'https://www.amazon.com/s?k=gap&page=3&crid=2HGTPE0NNQ3OO&qid=1696181153&sprefix=gap%2Caps%2C110&ref=sr_pg_4',
            'https://www.amazon.com/s?k=gap&page=3&crid=2HGTPE0NNQ3OO&qid=1696181153&sprefix=gap%2Caps%2C110&ref=sr_pg_5']

for url in url_list:
    driver.get(url)

    time.sleep(5)

    html = driver.page_source 

    soup = bs(html,'html.parser')

    bra_name = soup.find_all('span', class_="a-size-base-plus a-color-base")
    for x in bra_name:
        name = x.text
        print(name)
        brand_names.append(name)
    
    pro_name = soup.find_all('span', class_="a-size-base-plus a-color-base a-text-normal")
    for x in pro_name:
        name = x.text
        print(name)
        product_names.append(name)
    # Wait for the product ratings to be present on the page
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'a-size-base.puis-normal-weight-text')))
    pro_rate = soup.find_all('span', class_="a-size-base puis-normal-weight-text")
    for x in pro_rate:
        rate = x.text
        print(rate)
        product_ratings.append(rate)
    
    review_cou = soup.find_all('span', class_="a-size-base s-underline-text")
    for x in review_cou:
        review = x.text
        print(review)
        review_count.append(review)

    price = soup.find_all('span', class_="a-price-whole")
    for x in price:
        xx = x.text
        print(xx)
        product_price.append(xx)
```

```python
import csv
output_file = "Files_directory/method_2_GAP.csv"
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Brand Name', 'Product Name','Product Rating','Review Count','Product Price'])  # Write header row

    # Write data rows
    for i in range(len(review_count)):
        csv_writer.writerow([brand_names[i], product_names[i],product_ratings[i],review_count[i],product_price[i]])
```
